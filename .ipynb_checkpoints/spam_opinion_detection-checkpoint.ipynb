# Spam opinion 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('sum_data.csv')
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data = data.dropna()
data.shape
```

```python
data['len'] = data['reviewText'].str.count(' ')
data.head()
```

```python
df = data.drop(data[data.len < 15].index)
df.head()
```

```python
df2 = df.groupby(['reviewText']).size().to_frame('count')
df = pd.merge(df, df2, on=['reviewText'])
df
```

```python
spam_opinion = df[df['count'] > 1]
reviewText = spam_opinion.groupby(['reviewText', 'reviewerID',  'productID', ]).sum()
# df.to_csv('./spam_account.csv')
reviewText
```

```python
reviewerID = spam_opinion.groupby(['reviewerID', 'reviewText',  'productID', ]).sum()
reviewerID
```

```python
productID = spam_opinion.groupby(['reviewerID', 'reviewText',  'productID', ]).sum()
productID
```
